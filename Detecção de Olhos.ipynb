
# **Carregando a imagem**



```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
imagem = cv2.imread('imagem.jpg')
imagem.shape
```

```python
cv2_imshow(imagem)
```

```python
cv2.resize(imagem,(400,600));
```

```python
imagem_grey = cv2.cvtColor(imagem,cv2.COLOR_BGR2GRAY)
cv2_imshow(imagem_grey);
```

# **Detecção**

```python
detector_facial  = cv2.CascadeClassifier('/content/haarcascade_eye.xml')
```

```python
deteccoes = detector_facial.detectMultiScale(imagem_grey,scaleFactor=1.25)#scaleFactor é a escala da imagem sempre>1
deteccoes                                                #minNeighbors=x algoritmo escolhe entre x valores o melhor 
                                                         #minSize = tamanho minimo do segmento
                                                         #maxSize = tamanho maximo do segmento           
```

```python
for x,y,w,h in deteccoes:
  cv2.rectangle(imagem,(x,y),(x+h,y+h),(0,255,255))
cv2_imshow(imagem);
```
